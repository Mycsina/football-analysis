```python
import numpy as np
import pandas
import seaborn as sns

sns.set_theme(style="darkgrid")
```

```python
def money(string):
    clean = string[1:]
    if clean[-1] in "0123456789":
        return float(clean)
    elif clean[-1] == "K":
        return float(clean[:-1]) * 1000
    else:
        return float(clean[:-1]) * 1000000
```

```python
fifa17 = pandas.read_csv('../archive/FIFA17_official_data.csv')
```

```python
sns.lmplot(x="Age", y="Overall", data=fifa17.loc[fifa17["Best Position"] == "GK"],
                 x_estimator=np.mean, order = 2)
sns.lmplot(x="Age", y="Potential", data=fifa17.loc[fifa17["Best Position"] == "GK"],
                 x_estimator=np.mean, order = 2)
```

```python
fifa17["Value"] = fifa17["Value"].apply(money)
```

```python
sns.pairplot(fifa17[['Value', 'Overall', 'Potential', 'Age']], diag_kind='kde')
```
