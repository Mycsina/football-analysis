```python
import pandas as pd
import seaborn as sns

sns.set_theme(style="darkgrid")
```

```python
fifa17 = pd.read_csv('../archive/FIFA17_official_data.csv').convert_dtypes()
fifa17.drop(columns=["Best Overall Rating", "International Reputation", "Special"], inplace=True)
```

```python
# Every position
def pos_corr_matrix(fifa_dataset):
    pos_corr_matrix = {}
    for position in fifa_dataset["Best Position"].unique():
        pos_matrix = fifa_dataset.loc[fifa_dataset["Best Position"] == position]
        corr_matrix = pos_matrix.corr()
        pos_corr_matrix[position] = corr_matrix
    return pos_corr_matrix
```

```python
pos_corr_matrix = pos_corr_matrix(fifa17)
```

```python
# Most important stats for each position
for pos, corr_matrix in pos_corr_matrix.items():
    print(pos + " : " + corr_matrix["Overall"].drop(["Overall", "Potential"]).nlargest(10).to_string() + "\n")
```

```python
# Most important stats for each position's potential
for pos, corr_matrix in pos_corr_matrix.items():
    print(pos + " : " + corr_matrix["Potential"].drop(["Overall", "Potential"]).nlargest(10).to_string() + "\n")
```
